```python
# MEMBUAT MATRIKS 1 D
import numpy as np
A = np.array([0,0,1])
print("Matriks 1 Dimensi =", A)
```

```python
# MATRIK 2 D
B=np.array([[0,0,1],[1,1,2]])
B
```

```python
# MATRIKS SATUAN
SATU=np.ones((10,10))
# MATRIK NOL
NOL=np.zeros((5,5))
# MATRIKS RANDOM
RANDOM=np.random.random((7,7,))
print("Matriks satuan=", SATU)
print("Matriks nol=", NOL)
print("Matrik Random=", RANDOM)
```

```python
# MATRIKS KOSONG
Kosong=np.empty((2,2))
print("Matriks kosong=", Kosong)
# MATRIKS [5,10,15,....,100]
X=np.arange(5,105,5)
print("Matriks baru=", X)
# MATRIK DENGAN ELEMEN 0 - 1 SEBANYAK 10
Y=np.linspace(0,1,10)
print("Matriks Y=", Y)
```

```python
# OPERASI MATRIKS
p=np.array([[0,0,1],[1,1,0]])
q=np.array([[1,0,0],[1,1,1]])
# PENJUMLAHAN
J=np.add(p,q)
# PENGURANGAN
K=np.subtract(p.q)
print("p+q=", J)
print("p-q=", K)
```

```python
# BUATLAH MATRIKS BERUKURAN 2X2
A=np.array([[4,8],[8,16]])
B=np.array([[2,2],[4,4]])

# PERKALIAN AXB, np.multiply
AXB=np.multiply(A,B)

# PEMBAGIAN A:B, np.divide
AbagiB=np.divide(A,B)

# MENGHITUNG SISA BAGI
Sisa_Bagi=np.remainder(A,B)

print("AXB=", AXB)
print("A:B=", AbagiB)
print("Sisa hasil bagi=", Sisa_Bagi)


```
