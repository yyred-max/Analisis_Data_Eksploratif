Z test digunakan untuk jumlah data sosial media

```python
#48% oran tua mempercayai bahwa sosial media menyebabkan anak mereka stres

import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

n = 4500
pnull = 0.48
phat = 0.65
sm.stats.proportions_ztest(phat * n, n, pnull,alternative='larger')

#Karena nilai p value kurang dari H0, maka tolak H0
#Our calculated P-value of 1.2294951052777303e-126 is pretty small,
#and we can reject the null hypothesis, which is that social media is the cause of teenagers'
```
