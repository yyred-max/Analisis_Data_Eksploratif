```python
import pandas as pd
import numpy as np
df= pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv", delimiter=";")
df
```

```python
# Rata-rata
Rata_rata = df.mean()
print("Nilai rata-rata=", Rata_rata)
```

```python
#Median
Median = df.median()
print("Nilai tengah=", Median)
```

```python
#Modus
Modus = df.mode()
print("Nilai yang sering muncul=", Modus)
```

```python
df.describe()
```

```python
STD= df.std()
print("Nilai Standar Deviasi=", STD)
```

```python
Kemiringan=df.skew()
print("koefisien Skewness=", Kemiringan)
```

```python
Kurtosis = df.kurt()
print("Niali kurtosis=", Kurtosis)
```
